```python
feature_hospital_copy = feature_hospital_1
feature_hospital_copy = feature_hospital_copy.groupby('state_county').mean()
feature_hospital_copy['State_county']=feature_hospital_copy.index
feature_hospital_copy.reset_index(inplace=True)
feature_hospital_copy.drop('State_county',axis=1, inplace=True)
feature_hospital_copy['state_county'] = feature_hospital_copy['state_county'].apply(lambda x: str(x).rstrip())
```
